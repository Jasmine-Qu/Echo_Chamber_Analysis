```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
network_metrics_daily = pd.read_csv('/Users/chenyiqiu/Desktop/Semester 2/CS5344/Project/'
                         'data/processed data/2_processed data/network_metrics_daily.csv')
```

```python
network_metrics_daily.columns
```

```python
author_cnt = network_metrics_daily[['date', 'wallstreetbets_author_cnt', 'SecurityAnalysis_author_cnt',
       'investing_author_cnt', 'stocks_author_cnt', 'RobinHood_author_cnt']]
id_cnt = network_metrics_daily[['date', 'wallstreetbets_id_cnt', 'SecurityAnalysis_id_cnt', 'investing_id_cnt',
       'stocks_id_cnt', 'RobinHood_id_cnt']]
subscriber =network_metrics_daily [['date','wallstreetbets_subscribers_avg',
       'SecurityAnalysis_subscribers_avg', 'investing_subscribers_avg',
       'stocks_subscribers_avg', 'RobinHood_subscribers_avg']]
score =network_metrics_daily [['date','wallstreetbets_score_avg', 'SecurityAnalysis_score_avg',
       'investing_score_avg', 'stocks_score_avg', 'RobinHood_score_avg']]
comment =network_metrics_daily [['date','wallstreetbets_comment_avg', 'SecurityAnalysis_comment_avg',
       'investing_comment_avg', 'stocks_comment_avg', 'RobinHood_comment_avg']]
retention_rate =network_metrics_daily [['date','wallstreetbets_retention_rate', 'SecurityAnalysis_retention_rate',
       'investing_retention_rate', 'stocks_retention_rate',
       'RobinHood_retention_rate']]
connectivity = network_metrics_daily[['date','connectivity']]
```

```python
def plots(df,name):
    sns.set(style='white', palette = sns.color_palette("coolwarm",5))

    fig = plt.figure(figsize = (15,5))
    #fig, ax = plt.subplots(figsize = (15,5))
    for i in range(1,6):
        x = df[['date']]
        x = pd.to_datetime(x.date, format='%Y/%m/%d')
        y = df.iloc[:,i]
        ax = sns.lineplot(x = x,y = y, label=df.columns[i])
    
    ax.tick_params(axis='y',labelsize=13)
    ax.tick_params(axis='x',labelsize=13)
    ax.set_xlabel('date', fontsize=13)
    ax.set_ylabel('Count',fontsize=13)
    ax.set_title(name, fontsize=13) 
    sns.despine()
    
    return fig
```

```python
fig1 = plots(author_cnt,'author_cnt')
fig2 = plots(id_cnt,'id_cnt')
fig3 = plots(subscriber,'subscriber')
fig4 = plots(score,'score')
fig5 = plots(comment,'comment')
fig6 = plots(retention_rate,'retention_rate')
```

```python
sns.set(style='white', palette = sns.color_palette("coolwarm"))
fig7, ax = plt.subplots(figsize = (15,5))
x = connectivity[['date']]
x = pd.to_datetime(x.date, format='%Y/%m/%d')
y =  connectivity.iloc[:,1]
ax = sns.lineplot(x = x,y = y)
    
ax.tick_params(axis='y',labelsize=13)
ax.tick_params(axis='x',labelsize=13)
ax.set_xlabel('date', fontsize=13)
ax.set_ylabel('connectivity',fontsize=13)
ax.set_title('connectivity', fontsize=13) 
sns.despine()
```
